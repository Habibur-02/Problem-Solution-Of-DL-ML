```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# ১. মডেল কনফিগারেশন
model_id = "microsoft/Phi-4-mini-instruct"

print("Loading model... This might take a minute.")

# ২. মডেল লোড করা
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype=torch.float16,      # T4 GPU-এর জন্য float16 ই সেরা
    trust_remote_code=True,         # Phi-4 এর কাস্টম কোড চালানোর পারমিশন
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

# ৩. পাইপলাইন তৈরি
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

# ৪. চ্যাট ফরম্যাট
messages = [
    {"role": "system", "content": "You are a helpful AI assistant specialized in computer science."},
    {"role": "user", "content": "Explain the concept of 'Recursion' in programming with a simple Python example."}
]

# ৫. ইনফারেন্স রান
generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.7,
    "do_sample": True,
}

print("Generating response...")
output = pipe(messages, **generation_args)

print("-" * 50)
print(output[0]['generated_text'])
print("-" * 50)
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_id = "microsoft/Phi-3-mini-4k-instruct"

print(f"Loading {model_id} using native support...")

# ১. মডেল লোড করা
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=False,      # <--- পরিবর্তন: আমরা এখন বিল্ট-ইন সাপোর্ট ব্যবহার করছি
    _attn_implementation='eager'  # T4 GPU-এর জন্য নিরাপদ মোড
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

# ২. পাইপলাইন তৈরি
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

# ৩. প্রম্পট
messages = [
    {"role": "system", "content": "You are a helpful AI assistant."},
    {"role": "user", "content": "Explain the concept of 'Big O Notation' in one simple paragraph."}
]

# ৪. জেনারেশন কনফিগারেশন
generation_args = {
    "max_new_tokens": 300,
    "return_full_text": False,
    "temperature": 0.7,
    "do_sample": True,
}

print("Generating response...")
output = pipe(messages, **generation_args)

# ৫. আউটপুট
print("-" * 50)
print(output[0]['generated_text'])
print("-" * 50)
```

```python
def ask_phi(question):
    # নতুন প্রশ্নের জন্য মেসেজ ফরম্যাট করা
    messages = [
        {"role": "system", "content": "You are a helpful AI assistant."},
        {"role": "user", "content": question}
    ]

    # আগে থেকে লোড করা 'pipe' ব্যবহার করে উত্তর বের করা
    output = pipe(messages, **generation_args)
    return output[0]['generated_text']

# এখন তুমি ইচ্ছামতো প্রশ্ন করতে পারো
print("Question 1:")
print(ask_phi("What is Time Complexity?"))

print("\nQuestion 2:")
print(ask_phi("Write a python function to add two numbers."))
```

```python
print("Phi-3 এর সাথে চ্যাট শুরু হলো! (থামাতে 'exit' লিখুন)")

while True:
    user_input = input("\nতুমি: ")

    if user_input.lower() == 'exit':
        print("বিদায়!")
        break

    # উত্তর জেনারেট করা
    response = ask_phi(user_input)
    print(f"Phi-3: {response}")
```
